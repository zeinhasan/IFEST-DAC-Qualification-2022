```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Before Clustering

```python
c = pd.read_csv('Data - Kualifikasi.csv', delimiter = ";", decimal = ",")
c.fillna("Others", inplace=True)
c.head(2)
```

```python
plt.figure(figsize=(6,4))
ax = sns.countplot(data = c, x = "UKM", hue = "Fakultas", palette = "hls")
ax.set_title("UKM dan Fakultas")
ax.set_ylabel("Frekuensi")
ax.set_xlabel("")
sns.move_legend(ax, "lower center", bbox_to_anchor=(.5, -.3), ncol=3, title=None, frameon=False)
plt.savefig("Alamat dan Biaya.png")
plt.show()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(18, 18))
fig.suptitle('Visualisasi Data Mahasiswa', fontsize = 'xx-large', fontweight = 'bold')

#GENDER
p1 = sns.countplot(ax = axes[0,0], data = c, x = "Gender", palette = "hls")
p1.set_title("Gender")
p1.set_ylabel("Frekuensi")
p1.set_xlabel("")

#STATUS TINGGAL DENGAN
p2 = sns.countplot(ax = axes[0,1], data = c, x = "Tinggal_Dengan", palette = "hls")
p2.set_title("Status Tinggal Dengan")
p2.set_ylabel("Frekuensi")
p2.set_xlabel("")

#STATUS KERJA
p3 = sns.countplot(ax = axes[0,2], data = c, x = "Status_Kerja", palette = "hls")
p3.set_title("Status Kerja")
p3.set_ylabel("Frekuensi")
p3.set_xlabel("")

#BIAYA KULIAH
p4 = sns.countplot(ax = axes[1,0], data = c, x = "Biaya", palette = "hls")
p4.set_title("Biaya Kuliah")
p4.set_ylabel("Frekuensi")
p4.set_xlabel("")

#BIAYA KULIAH
p5 = sns.countplot(ax = axes[1,1], data = c, x = "Alamat", palette = "hls")
p5.set_title("Alamat")
p5.set_ylabel("Frekuensi")
p5.set_xlabel("")

#UKM
p6 = sns.countplot(ax = axes[1,2], data = c, x = "UKM", palette = "hls")
p6.set_title("UKM")
p6.set_ylabel("Frekuensi")
p6.set_xlabel("")

#ORGANISASI KAMPUS
p7 = sns.countplot(ax = axes[2,0], data = c, x = "Organisasi_Kampus", palette = "hls")
p7.set_title("Organisasi Kampus")
p7.set_ylabel("Frekuensi")
p7.set_xlabel("")

#FAKULTAS
p8 = sns.countplot(ax = axes[2,1], data = c, x = "Fakultas", palette = "hls")
p8.set_title("Fakultas")
p8.set_ylabel("Frekuensi")
p8.set_xlabel("")

#LAMA KULIAH
p9 = sns.boxplot(ax = axes[2,2], data = c, y = "Lama_Kuliah", palette = "hls")
p9.set_title("Lama Kuliah")
p9.set_ylabel("Tahun")
p9.set_xlabel("")
```

# After Clustering

```python
c1 = pd.read_excel('Cluster 1.xlsx')
c2 = pd.read_excel('Cluster 2.xlsx')
c3 = pd.read_excel('Cluster 3.xlsx')
c = pd.concat([c1, c2, c3])
c.shape
```

```python
c["Cluster label"].value_counts()
```

```python
c.head()
```

# Ringkasan

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
fig.suptitle('Visualisasi Hasil Clustering', fontsize = 'xx-large', fontweight = 'bold')

p1 = sns.countplot(ax = axes[0,0], data = c, x = "Cluster label", hue = "Status_Kerja", palette = "hls")
p1.set_title("Status Kerja")
p1.set_ylabel("Frekuensi")
p1.set_xlabel("")
sns.move_legend(p1, "lower center", bbox_to_anchor=(.5, -0.15), ncol=3, title=None, frameon=False)

p2 = sns.countplot(ax = axes[0,1],data = c, x = "Cluster label", hue = "Biaya", palette = "hls")
p2.set_title("Status Biaya Kuliah")
p2.set_ylabel("Frekuensi")
p2.set_xlabel("")
sns.move_legend(p2, "lower center", bbox_to_anchor=(.5, -0.15), ncol=3, title=None, frameon=False)

p3 = sns.countplot(ax = axes[1,0],data = c, x = "Cluster label", hue = "UKM", palette = "hls")
p3.set_title("Status UKM")
p3.set_ylabel("Frekuensi")
p3.set_xlabel("")
sns.move_legend(p3, "lower center", bbox_to_anchor=(.5, -0.25), ncol=3, title=None, frameon=False)

p4 = sns.countplot(ax = axes[1,1],data = c, x = "Cluster label", hue = "Organisasi_Kampus", palette = "hls")
p4.set_title("Status Organisasi Kampus")
p4.set_ylabel("Frekuensi")
p4.set_xlabel("")
sns.move_legend(p4, "lower center", bbox_to_anchor=(.5, -0.15), ncol=3, title=None, frameon=False)
```
